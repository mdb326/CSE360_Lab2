## Question 1: Free Fall Simulator

```python
def simulate(Δt, x, u, m=1, g=9.8):
    # Control input affects the acceleration
    dx = array([x[3], x[4], x[5], u[0]/m, u[1]/m, u[2]/m])
    
    # Include the gravity in the acceleration in z
    dx[5] += - g
        
    # Euler integration
    x += Δt * dx
    
    # Simulate the floor (Plastic collision)    
    if x[2] < 0:        
        x[2] = 0  # z = 0
        x[5] = 0  # vz = 0
    
    return x
```

```python
def control(t, y):    
    ux = 0
    uy = 0
    uz = 0  ### WRITE YOUR CONTROL POLICY HERE TO MAKE THE ROBOT FLY
    return [ux, uy, uz]
```

```python
tf = 3.
Δt = 0.1    # Time step
time = linspace(0.,tf, int(tf / Δt) + 1)  # Time interval


# Initial conditions
x = array([0., 0., 10., 0.,0.,0.])
x_log = [copy(x)]

for t in time:    
    y = sense(x)
    u = control(t, y)
    x = simulate(Δt, x, u)
    x_log.append(copy(x))
    
x_log = array(x_log)
```

## Question 3: Exercise 2 of Chapter 3, PD Controller

```python
def control2(t, y, d):    
    ux = 0
    uy = 0
    # zd - z + zvd - zv + mg
    uz = 3* (d[2] - y[2]) + 3* (0 - y[5])  + 1 * 9.8
    return [ux, uy, uz]
```

```python
tf = 3.
Δt = 0.1    # Time step
time = linspace(0.,tf, int(tf / Δt) + 1)  # Time interval


# Initial conditions
x = array([0., 0., 10., 0.,0.,0.])
x_log = [copy(x)]

for t in time:    
    y = sense(x)
    u = control2(t, y, array([0., 0., 15.]))
    x = simulate(Δt, x, u)
    x_log.append(copy(x))
    
x_log = array(x_log)
```
